# Getting dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# getting dataset
data_set1 = pd.read_csv('winemag-data-130k-v2.csv')
```

```python
data_set2 = pd.read_csv('winemag-data_first150k.csv')
data_set2.head()
```

# data correction and preprocessing

```python
data = pd.concat([data_set1, data_set2])
```

```python
data.head()
```

```python
data.shape
```

```python
data.info
```

```python
data.describe()
```

```python
data.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
data.head()
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
cols = ['region_1', 'region_2', 'taster_name', 'taster_twitter_handle', 'title',
       ]
data = data.drop(cols, axis = 1)
```

```python
data.isnull().sum()
```

Top 10 expensive wines
